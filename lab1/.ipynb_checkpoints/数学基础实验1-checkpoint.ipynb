首先是神经网络法求解。

# 导入所需库
`torch.nn` 包含着线性回归所需要的模型。\
`pandas` 用来处理数据。\
`matplotlib.pyplot` 用来将数据可视化。

```python
import torch
from torch import nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, Dataset
from pathlib import Path
from scipy.optimize import curve_fit

torch.__version__
```

# 处理数据
## 生成原函数和所需的数据
我所应用的原函数为$3x^2+5x+8$。
添加一定的，服从正态分布的噪声。

```python
# Define some hyperparameters of the original funciton
weight_square = 3
weight_linear = 5
weight_constant = 8

# Define the range of points to be generated
start = -100
end = 100
step = 2

# Get tensor
torch.manual_seed(42)
X = torch.arange(start, end, step).unsqueeze(dim=1)
y = weight_square * torch.square(X) + weight_linear * X + weight_constant + torch.randn(1)
```

## 将数据分为训练集和测试集
通常将全体数据的60～80%应用为训练集合。\
我在这之后也写了一个小的用于将点视觉化的函数。应用散点图。蓝点为训练集，绿色点为测试集合。

```python
# Create a split
train_split = (int)(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split] # To - :xxx
X_test, y_test = X[train_split:], y[train_split:] # From - xxx:

# Visualization
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
  """
  Visualize training data, test data and compare predictions.
  """
  plt.figure(figsize=(10, 7))
  plt.scatter(train_data, train_labels, c="b", s=4, label="Training data")
  plt.scatter(test_data, test_labels, c="g", s=4, label="Testing data")
  if predictions is not None:
    plt.scatter(test_data, predictions, c="r", s=4, label="Predictions")
  plt.legend(prop={"size" : 14});
  
plot_predictions()
```

# 编写线性回归模型
需要产生随机数，我采用`torch.manual_seed()`令产生的随机数相对固定，也就是说，在任何地方产生的随机数都是一样的，详情可以参考pyTorch的手册。

```python
# Create model class
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(42)
        self.weight_square = nn.Parameter(torch.randn(1,
                                                requires_grad=True,
                                                dtype=torch.float))
        self.weight_linear = nn.Parameter(torch.randn(1,
                                                     requires_grad=True,
                                                     dtype=torch.float))
        self.weight_constant = nn.Parameter(torch.randn(1,
                                                        requires_grad=True,
                                                        dtype=torch.float))
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weight_square * torch.square(x) + self.weight_linear * x + self.weight_constant
```

## 训练
### 设置Loss和优化器
由于这次拟合较为简单，可以将学习率设置低一些。

```python
# Instantiate the model
myModel = LinearRegressionModel()

# Setup a loss function
loss_fn = nn.L1Loss()

# Setup a optimizer
optimizer = torch.optim.SGD(params=myModel.parameters(),
                            lr=0.00001)
```

### 训练循环


```python
epochs = 300

epoch_count = []
loss_values = []
test_loss_values = []

for epoch in range(epochs):
  # Training Mode on
  myModel.train()

  # Forward
  y_pred = myModel.forward(X_train)

  # Loss
  loss = loss_fn(y_pred, y_train)
  # Optimize
  optimizer.zero_grad()

  loss.backward()

  optimizer.step()

  # Testing Mode on, Training Mode off
  myModel.eval()
  with torch.inference_mode():
    # Forward
    y_test_pred = myModel.forward(X_test)

    # Loss calculation
    test_loss = loss_fn(y_test_pred, y_test)
  if epoch % 200 == 0:
    print(f"Epoch: {epoch} | Loss: {loss} | Test Loss: {test_loss}")
  epoch_count.append(epoch)
  loss_values.append(loss)
  test_loss_values.append(test_loss)
```

```python
print("Parameters after training:")
print(myModel.state_dict())
print("\n")
with torch.inference_mode():
  plt.plot(epoch_count, loss_values, label="Loss")
  plt.plot(epoch_count, test_loss_values, label="Test Loss")
  plt.title("Training and test loss curves")
  plt.ylabel("Loss")
  plt.xlabel("Epochs")
  
  y_preds_new = myModel(X_test)
  plot_predictions(predictions=y_preds_new)

```

之后是公式法求解，应用拟合的算法。
调用`scipy.optimize.curve_fit()`求解。

```python
# Convert tensor to 1D np array
X_train_formula = X_train.numpy().flatten(order='C')
y_train_formula = y_train.numpy().flatten(order='C')
X_test_formula = X_test.numpy().flatten(order='C')
y_test_formula = y_test.numpy().flatten(order='C')


# Set the format of function
def target_func(x, a, b, c):
  return a*x**2 + b*x + c

popt, pcov = curve_fit(target_func, X_train_formula, y_train_formula)
print(popt)

y_preds_new = target_func(X_test_formula, *popt)
plot_predictions(predictions=y_preds_new)
```
